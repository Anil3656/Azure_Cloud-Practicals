```python
from azure.storage.blob import BlobServiceClient

# Initialize the BlobServiceClient
blob_account_url = "https://<your_storage_account>.blob.core.windows.net"
storage_credential = "Enter your storage account key here"
blob_url = "Your blob URL here"

blob_service_client = BlobServiceClient(blob_url=blob_url, credential=storage_credential)


#Download the blob data from Azure Blob Storage
with open("your_file_name", "rb") as download_file:
    download_file.write(blob_client.download_blob().readall())
    
    
#Anlyze the image using Document Intelligence
with open("your_file_name", "rb") as fd:
    poller = document_analysis_client.begin_analyze_document("prebuilt-document", document=fd)
    result = poller.result()
    
    # Print the extracted content from the document
    for idx, page in enumerate(result.pages):
        print("----Analyzing page #{}----".format(idx + 1))
        print("Page has width: {} and height: {}, measured with unit: {}".format(
            page.width,
            page.height,
            page.unit
        ))
        
        for line in page.lines:
            print("Line '{}' is within bounding box '{}'".format(
                line.content,
                line.bounding_box
            ))
```
