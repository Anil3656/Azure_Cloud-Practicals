# Create Container

```python
from azure.storage.blob import BlobServiceClient
```

# List out all container in your azure

# How to Delete the Container

```python
all_containers = Blob_service_client.list_containers()
for container in all_containers:
    print('Deleting Anilcontainer...')
    Blob_service_client.delete_container('Anilcontainer')
```

# Upload a blob

```python
blob_account_url = "Example: https://senapthiblobstorage.blob.core.windows.net/"
access_key = "Yoyr starage account access key"

Blob_service_client = BlobServiceClient(account_url=blob_account_url, credential=access_key)

Blob_service_client
```

```python
# Create a container
new_container = Blob_service_client.get_container_client('Newcontainer', public_access='container')
new_container 
```

# Uploading text File

```python
#Creating blobclient instance for specified bolb
blob_name = "Sample.txt"
file_path = "D:\Data Science Daily Updates\Senapthi_calss_Notes\AZURE LEARNINGS\Practicals\Sample.txt"
container_client = Blob_service_client.get_container_client('Newcontainer')
blob_client = container_client.get_blob_client(blob_name)

#Upload the file to the blob
with open(file_path,'rb') as data:
    blob_client.upload_blob(data, overwrite=True)
    
```
